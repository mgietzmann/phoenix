# 1. Load the Data

We're going to be loading a dataset that is a bunch of grayscale hand-drawn digits. And we're going to then try and train a model to recognize the numbers!

```python
import numpy as np
import keras
from keras import layers

num_classes = 10

(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

X_train.shape
```

# 2. The Simplest Model

Let's start with a model that has no hidden layers at all!

```python
input_shape = (28, 28, 1)
model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Flatten(),
        layers.Dense(num_classes, activation="softmax"),
    ]
)

model.summary()
```

Now let's setup our training loop.

```python
batch_size = 128
epochs = 15

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```

# 3. A "Linear" Model

Now let's add a hidden layer but leave it linear.

```python
input_shape = (28, 28, 1)
model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Flatten(),
        layers.Dense(64, activation="linear"),
        layers.Dense(num_classes, activation="softmax"),
    ]
)

model.summary()
```

```python
batch_size = 128
epochs = 15

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```

# 4. A Nonlinear Model

Let's use "ReLu" activation on our hidden layer.

```python
input_shape = (28, 28, 1)
model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Flatten(),
        layers.Dense(64, activation="relu"),
        layers.Dense(num_classes, activation="softmax"),
    ]
)

model.summary()
```

```python
batch_size = 128
epochs = 15

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```

That helped quite a bit!

# 5. More Nonlinearity!

Let's add another hidden layer.

```python
input_shape = (28, 28, 1)
model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Flatten(),
        layers.Dense(64, activation="relu"),
        layers.Dense(48, activation="relu"),
        layers.Dense(num_classes, activation="softmax"),
    ]
)

model.summary()
```

```python
batch_size = 128
epochs = 15

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```

So just adding complexity doesn't always help. Although we'll see later that if you add the *right* complexity it helps quite a lot.

# 6. The Relationship with Data

Okay so let's see what happens if we reduce the amount of data we have access to.

```python
n_samples = int(round(X_train.shape[0] / 10))  # Number of samples you want
indices = np.random.choice(len(X_train), size=n_samples, replace=False)
X_subset = X_train[indices]
y_subset = y_train[indices]
```

```python
input_shape = (28, 28, 1)
model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Flatten(),
        layers.Dense(64, activation="relu"),
        layers.Dense(48, activation="relu"),
        layers.Dense(num_classes, activation="softmax"),
    ]
)

model.summary()
```

```python
batch_size = 128
epochs = 15

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

model.fit(X_subset, y_subset, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```

Yikes! As always the amount of data you have is probably the most important aspect of building models.

# 7. The *Right* Complexity

What happens if we use a convolutional neural network?

```python
input_shape = (28, 28, 1)
model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax"),
    ]
)

model.summary()
```

Alright now if you try and train this model you'll find it is really really slow... this is where GPUs come in handy.

```python
batch_size = 128
epochs = 15

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```
